## **Grafico de datos**

```python
# Graficar
plt.figure(figsize=(10, 6))  # Tamaño de la figura
plt.plot(Data_Esperanza["Fecha"], Data_Esperanza["Esperanza de vida - Mujeres"], label="Esperanza de vida - Mujeres")
plt.plot(Data_Esperanza["Fecha"], Data_Esperanza["Esperanza de vida - Hombres"], label="Esperanza de vida - Hombres")
plt.plot(Data_Esperanza["Fecha"], Data_Esperanza["Esperanza de vida"], label="Esperanza de vida")

# Configurar título y etiquetas de los ejes
plt.title('Esperanza de Vida')
plt.xlabel('Fecha')
plt.ylabel('Esperanza de Vida')
plt.xticks(rotation=45)  # Rotar etiquetas del eje x para mayor claridad
plt.grid(True)  # Activar cuadrícula

# Mostrar la leyenda
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
